```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
income = pd.read_csv("data/income_statement_2017_2022.csv")
stud_hist = pd.read_csv("data/student_weeks_history.csv")

income.head(), stud_hist.head()
```

```python
income.describe(include="all")
stud_hist.describe()
```

```python
plt.figure()
plt.plot(income["Year"], income["TotalIncome"], marker="o")
plt.title("Total income by year")
plt.xlabel("Year")
plt.ylabel("€")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.plot(income["Year"], income["ProfitAfterTax"], marker="o")
plt.title("Profit after tax by year")
plt.xlabel("Year")
plt.ylabel("€")
plt.tight_layout()
plt.show()
```
